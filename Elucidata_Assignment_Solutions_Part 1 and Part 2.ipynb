```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline
```

### Reading the the Excel file which is made by manually copying the data from GCT file

```python
file_read = pd.read_excel("D:\General\Fall 2018_Spring 2019\Data Science\Jobs\Job Applications\Eulcidata\Assignment\PAAD.xlsx")
```

```python
file_read.head(2)
```

### Cleaning the data and slicing the Dataset of 18465 genes for 183 samples

```python
f2 = file_read.drop(labels=['Unnamed: 0' ], axis = 1 )

# last 18465 rows have data for all the Genes data
f3 = f2.tail(18465)
f3.head(5)
```

#### creating a dataset where rows represents samples and columns represents expression of genes

```python
f4 = f3.transpose()
f4.head(4)
```

```python
f5= f4.drop(labels='#1.3', axis = 0)
f5.head(2)
```

### Using PCA

Normalizing the data first before applying PCA

```python
# it is important to feature scale the data before using PCA
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
scaler.fit(f5)
```

```python
scaled_data = scaler.transform(f5)
```

```python
scaled_data
```

```python
f6 = pd.DataFrame(data = scaled_data)
f6.head(4)
```

#### We also need to do imputation for the missing values. Hence, we filled the mising values in the columns using the mean values

```python
f7 = f6.fillna(value = f6.mean() )
f7.head(6)
```

#### Applying PCA

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA(n_components=2)
```

```python
pca.fit(f7)
```

```python
# x_pca = X * u_reduce ( mxn * nxk )

x_pca = pca.transform(f7)
x_pca.shape
```

```python
pca.explained_variance_ratio_
```

## Solutions to Step A.

Run PCA on the dataset and plot different samples with PCA1 and PCA2 on x-axis andy-axis

```python
plt.figure( figsize= (12,8))

plt.scatter( x_pca[:,0] ,  x_pca[:,1] )
plt.xlabel('First principal component')
plt.ylabel('Second Principal Component')
```

## Solutions to Step B.

Overlaying the information from metadata column ‘histological_type_other’ on top of PCA plot and check if neuroendocrine tumors are separating out

#### Finding data on 'histological_type_other'

```python
cancer_type = f2.iloc[54]
cancer_type_a = cancer_type[1:]

cancer_type_b = list( cancer_type_a )
```

```python
cancer_type_a.unique()
```

#### Embedding 'histological_type_other' data in the dataframe and replacing the missing values with some other string

```python
f6['cancer_info'] = cancer_type_b
f6.head(2)
```

```python
f6['cancer_info'].fillna(value='Empty', inplace =True)
```

###  Creating a function to classify neuroendocrine tumors and applying it on the 'histological_type_other' to identify neuroendocrine tumors

```python
import re

def xop(var):
    
    if re.match("^neuroendocrine", var ) :
        
        return 1
    
         
    return 0    
```

```python
f6['cancer_marker'] = f6['cancer_info'].map(xop)
```

```python
f6.head()
```

```python
cancer_marker_list = list ( f6['cancer_marker'] )
len(cancer_marker_list)
```

## Plotting a final figure which has clearly seperates neuroendocrine tumors from the Adenocarcinoma samples

```python
plt.figure( figsize= (12,8))

plt.scatter( x_pca[:,0] ,  x_pca[:,1] , c= cancer_marker_list  , cmap='plasma' )
plt.xlabel('First principal component')
plt.ylabel('Second Principal Component')
```

● **What does the analysis say about the general behavior of the different samples?**

Answer: Analysis suggests that expression of many genes are correlated with each other and compressing the data for further processing makes sense, 

● **Are the neuroendocrine tumors clearly separable from the adenocarcinoma tumors?**

Answer: Yes, neuroendocrine tumors are clearly (visually) seperable. Although, to scientifically prove it we can use a classification algorithm.  

● **What can be said about the variance of the PCA?**

Answer: 2 components retain only about 25% (0.14 + 0.11) of variation/variance in the data.
